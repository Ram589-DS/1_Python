```python
#if statement --> Evaluates the condition and executes the block of code with in it,if the conditon is true.

age = 20
if age>=18:
    print("Age is greater than or equal to 18")
    
```

```python
age = int(input("Enter your age"))

if age>=18:
    print("Your are eligible to vote")
else:
    print("Your are not eligible to vote")
```

```python
# elif ---> allows you to check for multiple conditions . it stands for else if

age = int(input("Enter you age")) 

if age<13:
    print("You are a child")
elif age<18:
    print("You are teenager")
else:
    print("You are eligible to vote")
    
    

```

```python
# Nested Conditional statements

# You can place one or more if elif or else statements inside another if elif else statement to create nested conditional statements

# Ex : check if the number is even odd positive or negative

num = int(input("Enter your number"))

if num >=0:
    print("Number is positve num")
    if num%2 == 0:
        print("Number is even num")
    else:
        print("Number is odd number")
else:
    print("Number is negative num")
            
```
