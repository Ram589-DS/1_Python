```python
# for loop ---> it is used to iterate over the numbers

range(5)  # --> which means it gives you the range of 0 to 5 which is 0,1,2,3,4

# if you want to print the numbers from a particular range

# Syntax

for i in range(5):
    print(i)
```

```python
# for loop wrt Strings
str = "Ram Mohan"

# if i want to traverse through this particular string

for i in str:
    print(i)

```

```python
count = 0
while count < 5 :
    print(count)
    count = count+1;
```

```python
## loop control statements

# 1. break --> this statement exits the loop permanently
# 2. continue --> this skips the currrent iteration and contintues the with the next

for i in range(10):   # for i in range of 10 break the statement when i reaches to 5. till then print the i
    if(i==5):
        break
    print(i)




```

```python
# 2. continue --> this skips the currrent iteration and contintues the with the next

for i in range(10):
            if i%2==0:
                    continue    # if i is even num just continue then print the next num
            print(i)
# 
```

```python
# 3. pass statement --> it is a null operation.and it does nothing.

for i in range(5):
    if(i==3):
        print("Number is :" , i)  
        pass
    print(i)



 
```

```python
## Nested loops --> a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i : {i} and j : {j}")    # f string ---> f"" --> formating option of displaying both strings and values
        
```
