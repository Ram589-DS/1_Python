Variables 
Variables are fundemental elements in programming used to store data that can be reffered and manipulated in a program.In python variables are created when you assign a value to them,and they donot need explicit declaration to reserve memory space.The declaration happens automatically when you assign a value to the variable



```python
## Declaring and assigning variables.
age = 32
height = 5.7
name = "Ram"
a = 100 
is_student = True
# which means i created a variable age,height , name ,"a" , is_student and assign a value that particular variable. 

## printing the variables
print("Age : " , age)
print("Heght",height)
print("Name",name)
print("Value of A is : ",a)
print("Is he a student ? " , is_student)
```

```python
# Understanding variable types --> python is dynamically typed.type of a variable determined at the run time.
age = 25 # when i execute this during the run time it will understand that intiger is being assigned , it is an intiger variable.
height = 5.7 # float
name = "Ram" # string
is_student = True # boolean.

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))

```

```python
#Type Checking and Conversion
#type checking  --> if go ahead and write type(age) it tells the int. this is what type checking.
```

```python
# Can we convert one data type to another ?
age = 25
print(type(age))

# Type conversion

age_str = str(age) # str() is a function that converts the int to string

print(age_str)
print("Data type of the age_str is : ",type(age_str))
print("Data type of the age is :  " , type(age))
```

```python
# What all different type of Type conversions i can do ?

age = '25'
print(age)
print(type(age))
print(type(int(age))) # now i can convert that string to int.
# Some more examples

name = "Ram"
# can i convert this string to number which is int.
# int(name)
#same with float as well you cannot convert string to int or float.
```

```python
height = 5.7
type(height)

# now can i convert this float into intiger
int(height)
# now can i again convert this int to float

float(int(height))
```

```python
# Dynamic Typing --> python allows us to the type of the varaible to change as the program executes-> 

var = 10
print(var , type(var))

# Dynamic Typing --> python allows us to the type of the varaible to change as the program executes-> now i can go ahead and change the value of var 

var = "Hello"
print(var , type(var))

# again you can change this to some other data type

var = 5.1
print(var , type(var))

# so here python is allowing us to change the type of the variable while executing.
```

```python
# input -> let's say user wants to give some values,

age=input("what is your age")
print(age)
print(age , type(age)) # it gives you string.now you can convert this string to int by using the int function.

age = int(input("what is your age"))
print(age , type(age))
```
