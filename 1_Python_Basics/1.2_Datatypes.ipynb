DataTypes

1. Defintion


. Data types are a classification of data which tells the compiler or interpreter how the programmer intends to use that data
. They determine the type of operations that can be performed on the data , the values that data take and the amount of the memory


2. Importance of Data Types in Programming.

. Data types ensure that data is stored in a efficient way 
. It helps in performing correct operations on data
. Proper use of data types can prevent errors and bugs in the program
